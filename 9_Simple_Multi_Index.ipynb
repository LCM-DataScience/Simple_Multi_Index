```python
import pandas as pd
import numpy as np

# Lista de tickers do IBOV
tickers = ['PETR4', 'VALE3', 'ITUB4', 'BBDC4', 'B3SA3']

# Lista de datas
datas = ['2024-05-26', '2024-05-27']

# Gerando dados fictícios
np.random.seed(1)  # Para reprodutibilidade
data_spl_index = {
    'Ticker': np.tile(tickers, len(datas)),
    'Date': np.repeat(datas, len(tickers)),
    'Open': np.random.uniform(20, 100, size=len(tickers) * len(datas)),
    'Close': np.random.uniform(20, 100, size=len(tickers) * len(datas)),
    'Volume': np.random.randint(10000, 1000000, size=len(tickers) * len(datas))
}

# Criando o DataFrame com índice simples
df_spl_index = pd.DataFrame(data_spl_index)
df_spl_index = df_spl_index[['Date', 'Ticker', 'Open', 'Close', 'Volume']]

# Resetando o índice para ocultá-lo
df_spl_index = df_spl_index.reset_index(drop=True)

# Definindo o novo índice como "Date"
df_spl_index.index = df_spl_index['Date']

# Removendo a coluna duplicada "Date"
df_spl_index.drop(columns='Date', inplace=True)

# Exibindo o título
print("DataFrame com Índice Simples:")

# Exibindo o DataFrame final
df_spl_index
```

```python
# Lista de tickers do IBOV
tickers = ['PETR4', 'VALE3', 'ITUB4', 'BBDC4', 'B3SA3']

# Lista de datas
datas = ['2024-05-26', '2024-05-27']

# Lista de tipos de ações
tipos_act = ['Ordinária', 'Preferencial']

# Gerando dados fictícios
np.random.seed(1)
data_multi_index = {
    'Open': np.random.uniform(20, 100, size=len(datas)*len(tickers)*len(tipos_act)),
    'Close': np.random.uniform(20, 100, size=len(datas)*len(tickers)*len(tipos_act)),
    'Volume': np.random.randint(10000, 1000000, size=len(datas)*len(tickers)*len(tipos_act))
}

# Criando o MultiIndex com dados fictícios
index_levels = pd.MultiIndex.from_product([datas, tickers, tipos_act], names=['Date', 'Ticker', 'Tipo de ação'])

# Criando o DataFrame com MultiIndex
df_multi_index = pd.DataFrame(data_multi_index, index=index_levels)

# Exibindo o DataFrame com MultiIndex
print("DataFrame com MultiIndex:")
df_multi_index
```
